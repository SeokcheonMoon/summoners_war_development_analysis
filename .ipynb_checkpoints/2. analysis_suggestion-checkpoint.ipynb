```python
from konlpy.tag import Okt

# Okt 객체 생성
okt = Okt()

# 명사 추출 함수
def extract_nouns(text):
    nouns = okt.nouns(text)
    return nouns

# title 컬럼에서 명사 추출
df_suggestion['title_nouns'] = df_suggestion['title'].apply(extract_nouns)
df_suggestion
```
